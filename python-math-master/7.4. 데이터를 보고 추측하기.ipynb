### 7.4.1.	이동 평균

#### 코드 7-7 이동 평균 그래프 그리기

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 기온 데이터를 읽는다
dat= pd.read_csv('temperature.csv', encoding='UTF-8')

n = len(dat)       # 데이터 수
x = range(1, n+1)  # x축값（1～데이터 수）

# 기온
y = dat['평균기온']  # y축값（평균기온）
plt.plot(x, y)       # 그래프를 그린다

# 구간 수를 9로 지정한 이동평균
v = np.ones(9)/9.0
y2 = np.convolve(y, v, mode='same')  # 이동평균을 구한다
plt.plot(x[4:n-4], y2[4:n-4])        # 그래프를 그린다
plt.show()
```

### 7.4.2.	회귀직선

```python
import numpy as np
mean_x = np.mean(x)
mean_x
```

```python
mean_y = np.mean(y)
mean_y
```

```python
cov = np.mean((x-mean_x)*(y-mean_y))
cov
```

```python
var_x = np.var(x)
var_x
```

```python
a = cov / var_x
a
```

```python
b = mean_y - (a * mean_x)
b
```

#### 코드 7-8 회귀직선 그리기

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 데이터
x = np.array([23,24,28,24,27,21,18,25,28,20])  # 기온
y = np.array([37,22,62,32,74,16,10,69,83,7])   # 쥬스의 판매량

# 회귀직선
a, b = np.polyfit(x, y, 1)
y2 = a * x + b
print('기울기: {0}, 절편:{1}'.format(a, b))

# 그린다
plt.scatter(x, y)  # 산포도
plt.plot(x, y2)    # 회귀직선
plt.show()
```

```python
# 기온이 33도일 때 판매량을 예측한다
a * 33 + b
```
