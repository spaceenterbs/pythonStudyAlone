### 8.2.4.	미분 공식

#### 코드 8-3 함수와 도함수 그래프 그리기

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# x값
x = np.arange(-10, 10, 0.1)

# 본래의 함수f(x) = x**3 + 3x**2 + 3x + 1
y = x**3 + 3*x**2 + 3*x + 1
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()

# 도함수f'(x) = 3x**2 + 6x + 3
y2 = 3*x**2 + 6*x + 3
plt.plot(x, y2)
plt.grid(color='0.8')
plt.show()
```
