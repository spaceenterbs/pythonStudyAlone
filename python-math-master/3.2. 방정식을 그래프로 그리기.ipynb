#### 코드 3-2　y=3x-24 의 그래프

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 데이터
x = list(range(1,11))   # x의 값（1～10）
y = []
for i in range(10):
    y.append(3 * x[i] - 24)   # y = 3x - 24
    
# 그래프
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

### 3.2.3.	함수와 그래프

#### 코드 3-3　y=3x 의 그래프

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# 데이터
x = np.arange(-1.0, 1.01, 0.01)
y = 3 * x  # 일차함수

# 그래프를 그린다
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

#### y = x**2 의 그래프

```python
# 이차함수
y = x**2

# 그래프를 그린다
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

#### y = x**3 - x 의 그래프

```python
# 삼차함수
y = x**3 - x

# 그래프를 그린다
plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```
